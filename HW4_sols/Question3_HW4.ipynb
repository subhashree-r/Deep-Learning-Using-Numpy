```python
import sys
import codecs
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
import pickle
%matplotlib inline

from sklearn.manifold import TSNE
########################## Reading the words and the corresponding vectors #############################
def main():
    f= open('vectors_new.txt',"r")
    lines = f.readlines()
    words = []
    vectors =[]
    for l in lines:
#         print("loop")
        arr = l.split(' ')
        vec = arr[1:]
        vec = [float(i) for i in vec]
        word = str(arr[0])
        words.append(word)
        vectors.append(vec)
    with open("words.txt", "wb") as fp:   #Pickling
        pickle.dump(words, fp)
    with open("vector.txt","wb") as fp1:
        pickle.dump(vectors,fp1)

if __name__=='__main__':
    main()
```

```python
import pickle
with open("words.txt", "rb") as fp:   # Unpickling
    wor = pickle.load(fp)

with open("vector.txt","rb") as fp1:
    vec = pickle.load(fp1)
    
    
print(len(wor))
print(len(vec))
```

The below code does retrieval of closest 20 words using cosine similarity

```python
#word = life
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
ind = wor.index("life")
v1 = vec[ind]
print(ind,v1)
print(len(v1),len(vec[2]))
print(cosine_similarity([v1], [vec[2]]))
# print()
wor_vec={}
for i in range(len(vec)) :
    c = cosine_similarity([v1],[vec[i]])
    wor_vec[wor[i]]=abs(c[0][0])
    
v=sorted(wor_vec.values())
k=sorted(wor_vec,key=wor_vec.get)

print(k[1:20])
    
    
```
